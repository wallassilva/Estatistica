## Probabilidade

1- A pesquisa sobre filmes mostrou que 50% das pessoas gostam dos filmes da Marvel, 40% dos filmes da DC. Além disso mostrou que 30%
gostam ao mesmo tempo dos filmes da Marvel e da DC. Qual a probabilidade de uma pessoa escolhida ao acaso da população não goste nem da Marvel e nem da DC ?

```python
# Probabilidade de gostar da Marvel
prob_marvel= 0.5

# Probabilidade de gostar da DC
prob_dc = 0.4

# Probabilidade de gostar da Marvel e da DC ao mesmo tempo
prob_marvel_dc = 0.3

# Probabilidade de não gostar da Marvel e nem da DC
prob_nao_gosta = 1 - (prob_marvel + prob_dc - prob_marvel_dc)

print("Probabilidade de não gostar da Marvel e nem da DC:", round(prob_nao_gosta, 2))
```

3- Dois dados não viciados são lançados e são somados os números das faces voltadas para cima. Qual aprobabilidade de que sua soma seja 9 ?

```python
#Crinado todas as combinações possíveis para o problema
combinacoes = [(i, j) for i in range(1, 7) for j in range(1, 7)]

# contando o número de combinações cuja soma é igual a 9
soma_nove = 0
for (dado1, dado2) in combinacoes:
    if (dado1 + dado2) == 9:
        soma_nove += 1

# calculando a probabilidade de que a soma seja igual a 9
prob = soma_nove / len(combinacoes)

print("A probabilidade de que a soma seja igual a 9 é:", "{:.2f}".format(prob))
```

##Variáveis aleatórias discretas

### Distribuição bernoulli
1- Suponha que a probabilidade de óbito de um paciente, ao dar entrada na terapia intensiva, seja de 30% (risco de morte). Qual a probabilidade desse paciente se recuperar?

```python
# Distribuição bernoulli

from scipy.stats import bernoulli

p = 0.3  # probabilidade de óbito
X = bernoulli(p)

Prob_recuperar = X.pmf(0)  # probabilidade de sobrevivência (X = 0)

print("A probabilidade do paciente se recuperar é de", Prob_recuperar)
```

### Distribuição binominal

2- Em uma prova objetiva realizada em sala de aula, com 10 questôes e cada uma com 5 alternativas, , qual a probabilidade de acertar 7 questões no chute ?

```python
# Distribuição binominal

from scipy.stats import binom

q = 10 # número de tentativas
p = 1/5 # probabilidade de acertar uma questão
k = 7 # número de sucessos desejado
vabin = binom(q, p)
prob = vabin.sf(k - 1)
print(prob)

```

### Distribuição poison
3- Em uma rua movimentada, costuma-se passar 50 pessoas caminhando por hora, qual a possibilidade de passar 5 ou pessoas na mesma rua em um determidado minuto?


```python
# Distribuição poison

from scipy.stats import poisson

lambda_valor = 50/60  # conversão de carros por hora para carros por minuto
k = 3  # número de carros esperados em um minuto

#Funçao que calcula a probabilidade de 5 ou mais pessoas passando na rua
probb = poisson.sf(k-1, lambda_valor)
print(f"A probabilidade de {k} ou mais pessoas passarem pela rua em um minuto é {probb:.4f}")
```

### Distribuições Uniforme
1- José é um estudante e todos os dias vai de ônibus para a faculdade. Há apenas um ônibus que te serve e ele passa no ponto em qualquer instante entre as 12H e 12H 30 minutos. Se José chegar ao ponto às 12H 24 minutos, qual a probabilidade de que ele consiga pegar o ônibus ?

```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt

a, b = 0, 30

# Definindo a distribuição uniforme contínua entre 12h e 12h30min
dist = uniform(a, b)

# Probabilidade de que o ônibus passe depois que José chega ao ponto às 12h24min
prob = dist.cdf(b) - dist.cdf(24)

# CDF da distribuição uniforme
x = np.linspace(a, b, 1000)
cdf = dist.cdf(x)

dist_prob = dist.pdf(x)


plt.subplots(figsize=(14,6))
plt.plot(x,dist_prob, color='red')
x_points = [0, 0]
y_points = [0, 1/30]
plt.plot(x_points, y_points, linestyle = 'dashed',color='green')
x_points = [6, 6]
y_points = [0, 1/30]

plt.xlabel('x')
plt.ylabel('Probabilidade')
plt.title('Distribuição uniforme entre 0 e 30 minutos')
plt.grid(True)
plt.show()


# Plot do gráfico
fig, ax = plt.subplots(figsize=(14, 6))

ax.plot(x, cdf, color='red')
ax.set_xlabel('Tempo (minutos após 12h)')
ax.set_ylabel('Probabilidade acumulada')
ax.set_title('CDF da distribuição uniforme entre 12h e 12h30min')
ax.grid(True)
ax.axvline(x=24, linestyle='dashed', color='green')
ax.axhline(y=dist.cdf(24), linestyle='dashed', color= 'green')
ax.text(24, 0.05, f'Probabilidade = {prob:.2f}', ha='right', va='bottom')
plt.show()

```

### Distribuições Normal
3- Suponha que o tempo necessário para atendimento de clientes em uma central de atendimento telefônico siga uma distribuição normal de média de 8 minutos e desvio padrão de 2 minutos. Qual é a probabilidade de que um atendimento dure menos de 5 minutos?

```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt

media = 8  # média
desvio_padrao = 2  # desvio padrão
tempo = 5  # tempo em minutos

# Calculando a pontuação z
z = (tempo - media) / desvio_padrao
prob = st.norm.cdf(z)
# Criando o eixo x
x = np.linspace(media - 3*desvio_padrao, media + 3* desvio_padrao, 1000)

# Calculando a densidade de probabilidade para cada valor de x
y = st.norm.pdf(x, loc=media, scale= desvio_padrao)

# Criando o gráfico
plt.plot(x, y, color='blue')
plt.axvline(x=tempo, linestyle='--', color='red')
plt.xlabel('Tempo (minutos)')
plt.ylabel('Densidade de probabilidade')
plt.title('Distribuição normal do tempo de atendimento')
plt.show()
print(f"A probabilidade de um atendimento durar menos de {tempo} minutos é de {prob:.2%}")
```
