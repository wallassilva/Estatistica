```python
#Frequência das equipes participantes da temporada(2010-2021)
frequencia = equipes.team.value_counts()
frequencia
```

```python
#Transformando uma serie em DataFrame
frequencia = pd.DataFrame(frequencia)
frequencia.reset_index(level=0, inplace=True)
frequencia
# Renomeando os nomes da colunas
frequencia.rename(columns={'index':'Equipes', 'team':'Participacoes'},inplace=True)
frequencia
```

# Análise Exploratória - Graficos Sobre as Equipes





```python
import seaborn as sns; sns.set();
import matplotlib.pyplot as plt;
import plotly.express as px
```

```python
plt.figure(
    figsize=(20,10),
     )

Ax = plt.axes()
Ax.set_facecolor('white') # Definindo cor de fundo no gráfico

plt.xticks(rotation=90)
plt.title("Quantidade de pontos durante o período (2010-2021)")

for i in range(len(soma1)):
    plt.annotate(str(round(soma1['Pontos'][i],2)),xy=(soma1.index[i], soma1['Pontos'][i]), ha='center', va='bottom')

sns.barplot(x='Equipes',y='Pontos',data=soma1) #palette='Dark2'
plt.ylabel("Total de Pontos", # nome do label
           fontsize='large', # Tamanho da fonte
            color='blue' # Cor da Fonte)
            )
plt.xlabel("Equipes Participantes",  # nome do label
           fontsize='large', # Tamanho da fonte
            color='blue' # Cor da Fonte)
            )
plt.title("Quantidade de pontos durante o período (2010-2021)")
plt.tight_layout();
```

O gráfico mostra a quantidade de pontos de cada equipe, sendo os pontos uma soma dos pontos conquistado nas temporadas (2010-2021)

```python
plt.figure(figsize=(20,10))
Ax = plt.axes()
plt.title("Frequencia de cada equipe participantes durante o período (2010-2021)")

Ax.set_facecolor('white') # Definindo cor de fundo no gráfico
plt.xticks(rotation=90) # Rotacionando legenda

for i in range(len(frequencia)):
 plt.annotate(str(round(frequencia['Participacoes'][i], 2)), xy = (frequencia.index[i], frequencia ['Participacoes'][i]), ha='center', va='bottom')

sns.barplot(x='Equipes',y= 'Participacoes',data=frequencia) #palette='Dark2'

plt.xlabel("Equipes Participantes", # nome do label
           fontsize='large', # Tamanho da fonte
            color='blue') # Cor da Fonte)
plt.ylabel("Participacoes por temporada",
           fontsize='large',
            color='blue'
           )

plt.tight_layout();
```


O gráfico mostra a quantidade de Participações de cada equipe durante as temporadas (2010-2021)

```python
nacionalidade = pilotos.nationality
nacionalidade
repeticao = nacionalidade.value_counts()
repeticao
```

```python
repeticao = pd.DataFrame(repeticao)
repeticao.reset_index(level=0, inplace=True)
repeticao
# Renomeando os nomes da colunas
repeticao.rename(columns={'index':'Paises', 'nationality':'Representantes'},inplace=True)
repeticao
```

```python
pais_predominante = pilotos.groupby('nationality')[['points']].sum().reset_index()
pais_predominante.columns
```
