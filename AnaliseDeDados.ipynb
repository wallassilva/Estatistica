## Relatório referente a base de dados de Equipes

```python
equipes
```

```python
# Verificando os tipos do Banco de dados
dict(equipes.dtypes)
```

```python
# Soma de pontos de cada equipe (2010-2021)
soma1 = equipes.groupby(['team'])['points'].sum()
soma1
```

```python
#Frequência das equipes participantes da temporada(2010-2021)
frequencia = equipes.team.value_counts()
frequencia
```

```python
#Transformando uma serie em DataFrame
frequencia = pd.DataFrame(frequencia)
frequencia.reset_index(level=0, inplace=True)
frequencia
# Renomeando os nomes da colunas
frequencia.rename(columns={'index':'Equipes', 'team':'Participacoes'},inplace=True)
frequencia
```

```python
#Transformando uma serie em DataFrame
soma1 = pd.DataFrame(soma1)
soma1.reset_index(level=0, inplace=True)

# Renomeando os nomes da colunas
soma1.rename(columns={'points':'Pontos', 'team':'Equipes'},inplace=True)
soma1
```

## Relatório referente a base de dados de Pilotos

```python
pilotos
```

```python
nacionalidade = pilotos.nationality
nacionalidade
repeticao = nacionalidade.value_counts()
```

```python
repeticao.describe()
```

```python
# Transformando uma Serie em um DataFrame
repeticao = pd.DataFrame(repeticao)
repeticao.reset_index(level=0, inplace=True)
repeticao
# Renomeando os nomes da colunas
repeticao.rename(columns={'index':'Paises', 'nationality':'Representantes'},inplace=True)
repeticao
```

```python
pais_predominante = pilotos.groupby('nationality')[['points']].sum().reset_index()
pais_predominante.columns
```

```python
pais_predominante.groupby('points').max().sort_values(by = 'points',ascending = 0)
```

```python
round(pais_predominante[['points']].describe())
```

## Análise Exploratória - Graficos Sobre as Equipes





```python
import seaborn as sns; sns.set();
import matplotlib.pyplot as plt;
import plotly.express as px
```

```python
plt.figure(
    figsize=(16,10),
     )

Ax = plt.axes()
Ax.set_facecolor('white') # Definindo cor de fundo no gráfico

plt.xticks(rotation=90) # Rotacionando legenda
plt.title("Quantidade de pontos durante o período (2010-2021)", # Título da tabela
                 fontsize='large', # Tamanho da fonte
            color='blue') # Cor da Fonte

for i in range(len(soma1)):
    plt.annotate(str(round(soma1['Pontos'][i],2)),xy=(soma1.index[i], soma1['Pontos'][i]), ha='center', va='bottom')

sns.barplot(x='Equipes',y='Pontos',data=soma1)

plt.ylabel("Total de Pontos", # nome do label
           fontsize='large', # Tamanho da fonte
            color='blue' # Cor da Fonte
            )
plt.xlabel("Equipes Participantes",  # nome do label
           fontsize='large', # Tamanho da fonte
            color='blue' # Cor da Fonte
            )

plt.tight_layout();
```

O gráfico mostra a quantidade de pontos de cada equipe, sendo os dados uma soma dos pontos conquistado nas temporadas (2010-2021)

```python
plt.figure(figsize=(10,8))
Ax = plt.axes()
plt.title("Frequencia de cada equipe participantes durante o período (2010-2021)", # Título da tabela
            fontsize='large', # Tamanho da fonte
            color='blue') # Cor da Fonte

Ax.set_facecolor('white') # Definindo cor de fundo no gráfico
plt.xticks(rotation=90) # Rotacionando legenda

for i in range(len(frequencia)):
 plt.annotate(str(round(frequencia['Participacoes'][i], 2)), xy = (frequencia.index[i], frequencia ['Participacoes'][i]), ha='center', va='bottom')

sns.barplot(x='Equipes',y= 'Participacoes',data=frequencia)
plt.xlabel("Equipes Participantes", # nome do label
           fontsize='large', # Tamanho da fonte
            color='blue') # Cor da Fonte)
plt.ylabel("Participacoes por temporada",
           fontsize='large',
            color='blue'
           )

plt.tight_layout();
```


O gráfico mostra a quantidade de Participações de cada equipe durante as temporadas (2010-2021)

## Análise Exploratória - Graficos Sobre os Pilotos

```python
plt.figure(figsize=(10,8))
Ax = plt.axes()
plt.title('Nacionalidade dos pilotos participantes durante o período (2010-2021)',
         fontsize='large', # Tamanho da fonte
            color='blue') # Cor da Fonte))

Ax.set_facecolor('white')

for i in range(len(repeticao)):
 plt.annotate(str(round(repeticao['Representantes'][i], 2)), xy = (repeticao.index[i], repeticao ['Representantes'][i]), ha='center', va='bottom')

sns.barplot(x='Paises',y= 'Representantes',data=repeticao,palette='Dark2')

plt.xlabel("Países Participantes", # nome do label
           fontsize='large', # Tamanho da fonte
            color='blue') # Cor da Fonte)
plt.ylabel("Quantidade de pilotos por país",
           fontsize='large',
            color='blue'
           )

plt.tight_layout();

```

O gráfico mostra a Nacionalidade dos pilotos participantes durante o período (2010-2021).


```python
plt.figure(figsize=(15,8))
Ax = plt.axes()
plt.title("Quantidade de pontos de cada país no período de (2010-2021)",
          fontsize='large', # Tamanho da fonte
            color='blue') # Cor da Fonte))

Ax.set_facecolor('white') # Definindo cor de fundo no gráfico
plt.xticks(rotation=90) # Rotacionando legenda

for i in range(len(pais_predominante)):
 plt.annotate(str(round(pais_predominante['points'][i], 2)), xy = (pais_predominante.index[i], pais_predominante ['points'][i]), ha='center', va='bottom')

sns.barplot (x='nationality',y= 'points',data=pais_predominante) #palette='Dark2'

plt.xlabel("Paises Participantes", # nome do label
           fontsize='large', # Tamanho da fonte
            color='blue') # Cor da Fonte)
plt.ylabel("Pontos",
           fontsize='large',
            color='blue'
           )

plt.tight_layout();
```

O gráfico mostra a quantidade de pontos conquistados por pilotos do mesmo país.
